```python
def get_statewie_data(by_demographics=False):
    if by_demographics:
        return inventory[~(inventory["Locale"].str.contains("-"))
                         & ~(inventory["Timeframe"].str.contains("-"))
                         & ~(inventory["Dimension List"] == "Total")][["Timeframe",
                                                                       "Dimension List",
                                                                       "Numeric Value","Locale"]]    
    else:
        return inventory[inventory["Locale"].str.contains("-")
                         & ~(inventory["Timeframe"].str.contains("-"))
                         & (inventory["Dimension List"] == "Total")][["Timeframe","Numeric Value","Locale"]]


def get_data(by_demographics=False):
    if by_demographics:
        return inventory[inventory["Locale"].str.contains("-")
                         & ~(inventory["Timeframe"].str.contains("-"))
                         & ~(inventory["Dimension List"] == "Total")][["Timeframe",
                                                                       "Dimension List",
                                                                       "Numeric Value","Locale"]]    
    else:
        return inventory[inventory["Locale"].str.contains("-")
                         & ~(inventory["Timeframe"].str.contains("-"))
                         & (inventory["Dimension List"] == "Total")][["Timeframe","Numeric Value","Locale"]]
    
get_statewie_data(by_demographics=True)
```

```python
demographics_county = get_statewie_data(by_demographics=True)
demographics_county["Numeric Value"] = pd.to_numeric(demographics_county["Numeric Value"],errors="coerce")


race = demographics_county[~demographics_county["Dimension List"].str.contains("Aged")]
race = pd.pivot_table(race,index="Dimension List",
               columns="Timeframe",
               values="Numeric Value").transpose()
race
```

```python
age = demographics_county[demographics_county["Dimension List"].str.contains("Aged")]
age = pd.pivot_table(age,
                     index="Dimension List",
                     columns="Timeframe",
                     values="Numeric Value").transpose()
age
```
